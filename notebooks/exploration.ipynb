## 1  Load Raw Data

```python
# TODO: update this path to point to your downloaded corpus
RAW_FILE = Path('../data/raw/input.txt')

if RAW_FILE.exists():
    with open(RAW_FILE, 'r', encoding='utf-8') as f:
        text = f.read()
    print(f'Loaded {len(text):,} characters from {RAW_FILE}')
    print('First 200 characters:')
    print(text[:200])
else:
    print(f'File not found: {RAW_FILE}')
    print('See data/README.md for download instructions.')
    text = 'Hello, World! This is a placeholder.'  # use dummy text for exploration
```

## 2  Basic Statistics

```python
# TODO: compute and print basic statistics about the corpus
total_chars = len(text)
unique_chars = sorted(set(text))
vocab_size = len(unique_chars)

print(f'Total characters  : {total_chars:,}')
print(f'Unique characters : {vocab_size}')
print(f'Unique char list  : {unique_chars}')
```

## 3  Character Frequency Distribution

```python
# TODO: plot character frequencies
char_freq = Counter(text)
chars, counts = zip(*sorted(char_freq.items(), key=lambda x: -x[1]))

plt.figure(figsize=(14, 4))
plt.bar(range(len(chars)), counts)
plt.xticks(range(len(chars)), [repr(c) for c in chars], rotation=90, fontsize=7)
plt.title('Character Frequency Distribution')
plt.xlabel('Character')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
